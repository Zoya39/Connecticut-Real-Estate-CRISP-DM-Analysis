### 05-2 Exploring Real Estate trends by Sales Amount

```python
grouped_df = df_sold.groupby('Property Type')['Sale Amount'].sum().reset_index()
```

```python
grouped_df
```

#### - by Property Type

```python
# Creating the bar plot
bar2=df_sold.groupby('Property Type')['Sale Amount'].sum().sort_values(ascending=False).plot.barh()
# Adding title and labels
plt.title('Sum of Sale Amount by Property Type')
plt.xlabel('Property Type')
plt.ylabel('Sum of Sale Amount')
```

```python
### Saving plot:
bar2.figure.savefig(os.path.join(path, 'Analytics', 'SumSalesPropertyType_barplot.png'))
```
